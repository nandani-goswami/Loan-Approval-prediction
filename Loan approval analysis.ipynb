```python
def plot_cdf(data, bin_size,title=None, xlabel=None):

    counts, bin_edges = np.histogram (data, bins=bin_size, normed=True)
    cdf = np.cumsum(counts)
    
    sns.set_style("whitegrid")
    plt.figure(figsize=(18,13))
    plt.plot (bin_edges[1:], cdf/cdf[-1])
    plt.ylabel('CDF')
    
    if title:
        plt.title(title)
    if xlabel:
        plt.xlabel(xlabel)
    plt.show()
```

```python
plot_cdf(df['loan_amnt'],20,'loan Amount','Amount')#ploting the cdf of amount loan taken by Customer's
```

```python
"""Conculsion

CDF Plot :-
Y Axix:- Tell us about Percentage
X Axis:-Tell about Amount
For Example:- about 78% customer of 15000 loan amount"""
```
